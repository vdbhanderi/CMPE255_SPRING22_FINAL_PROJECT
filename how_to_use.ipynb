```python
from tensorflow import keras
model = keras.models.load_model('/content/drive/MyDrive/255/project/char_level_accuracy_84_3lstm_rnn_8inputlength-20220503T204129Z-001/char_level_accuracy_84_3lstm_rnn_8inputlength')
```

```python
max_length = 74
```

```python
#change the string with a specifc input from the daatset.
# only work on the data from the dataset
test_string = ["wadeye residents eskeing return to outstaitosn	"]

test_sample = g_text_tokenizer.texts_to_sequences(test_string)
test_sample = pad_sequences(test_sample, maxlen=max_length, padding='post')
```

```python
pre = model.predict(test_sample)
def deTokenize(logits):
    index_to_words = {id: word for word, id in text_tokenizer.word_index.items()}
    index_to_words[0] = ''
    pre_index = np.argmax(logits, 1)
    return ''.join(index_to_words[prediction] for prediction in pre_index)

print("Iutput:    ",test_string[0])
print("Predicted: ",deTokenize(pre[0]))
```
