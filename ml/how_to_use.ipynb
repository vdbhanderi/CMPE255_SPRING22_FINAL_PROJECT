```python
import tensorflow as tf
import numpy as np
from keras.preprocessing.sequence import pad_sequences
```

```python
import json
with open('text_tokenizer.json') as f:
    data = json.load(f)
    text_tokenizer = tf.keras.preprocessing.text.tokenizer_from_json(data)

with open('g_text_tokenizer.json') as f:
    data = json.load(f)
    g_text_tokenizer = tf.keras.preprocessing.text.tokenizer_from_json(data)
```

```python
from tensorflow import keras
model = keras.models.load_model('char_level_accuracy_84_3lstm_rnn')
```

```python
max_length = 74
```

```python
#change the string with a specifc input from the daatset.
# only work on the data from the dataset
test_string = ["wadeye residents eskeing return to outstaitosn"]

test_sample = g_text_tokenizer.texts_to_sequences(test_string)
test_sample = pad_sequences(test_sample, maxlen=max_length, padding='post')
```

```python
pre = model.predict(test_sample)
def deTokenize(prediction):
    index_to_words = {id: word for word, id in text_tokenizer.word_index.items()}
    index_to_words[0] = ''
    pre_index = np.argmax(prediction, 1)
    return ''.join(index_to_words[word] for word in pre_index)

print("Iutput:    ",test_string[0])
print("Predicted: ",deTokenize(pre[0]))
```
